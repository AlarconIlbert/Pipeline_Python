```python
import pandas as pd
```

```python
datos=pd.read_csv('/content/poblacion.csv')
```

```python
datos.rename(columns={'Y':'Year','Estados Unidos':'EEUU'},inplace=True)
```

```python
datos.drop(columns='Nigeria',inplace=True)
```

```python
datos.head(5)
```

```python
import sqlite3 as sql
```

```python
conex=sql.connect("poblacion.db")
```

```python
conex.commit()
```

```python
cursor=conex.cursor()
```

```python
cursor.execute("""CREATE TABLE IF NOT EXISTS habitantes (Year INT, China real, India real, EEUU real, Indonesia real, Pakistan real, Brasil real)""")
conex.commit()
```

```python
cursor.execute("INSERT INTO habitantes (Year, China, India, EEUU, Indonesia, Pakistan, Brasil) VALUES(1910,448.77,253.1,117.1,78.86,27.46,43.15)")
conex.commit()
```

```python
datos.to_sql('habitantes', conex, if_exists='replace',index=False)
```

```python
cursor.execute("SELECT * FROM habitantes")
```

```python
rows = cursor.fetchall()
```

con fetch all obtenemos una lista de tuplas

```python
for row in rows:
  print(row)
```

```python
cursor.execute("SELECT China FROM habitantes WHERE Year = 1950")
```

```python
resultado = cursor.fetchall()
```

```python
for result in resultado:
  print(result[0])
```
